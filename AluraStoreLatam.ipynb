# 4. Productos más y menos vendidos

**Introducción**

El presente informe tiene como propósito determinar cuál de las cuatro tiendas debería ponerse en venta, a partir del análisis comparativo de su desempeño en distintos indicadores clave: ingresos totales, categorías de productos más vendidas, calificaciones promedio de los clientes, productos más y menos vendidos, y costes de envío promedio. **Metodología**

Datos analizados: ventas totales, calificaciones promedio, categorías predominantes, productos más y menos vendidos, y costes de envío en las cuatro tiendas.

Ingresos totales (indicador de rentabilidad y volumen de ventas).

Diversidad y rotación de productos (indicador de vitalidad comercial).

**Criterio de decisión:** La tienda con menor desempeño global —es decir, menores ingresos, menor calificación, y peores condiciones logísticas— será la candidata ideal para ser vendida.
